```python
access_to_rural = path+'access_to_elec_rural__world/API_EG.ELC.ACCS.RU.ZS_DS2_en_csv_v2_4364070.json'
access_to_urban = path+'access_to_electricity_urban__world/API_EG.ELC.ACCS.UR.ZS_DS2_en_csv_v2_4367536.json'
access_to_world = path+'access_to_electricity__total/API_EG.ELC.ACCS.ZS_DS2_en_csv_v2_4353549.json'
transmission_distribution_losses = path+'electricity_power_transmissoin_distribution_losses/API_EG.ELC.LOSS.ZS_DS2_en_csv_v2_4357259.json'
production_by_nuclear = path+'electricity_production_nuclear_sources/API_EG.ELC.NUCL.ZS_DS2_en_csv_v2_4357319.json'
production_by_oil = path+'electricity_production_oil_sources/API_EG.ELC.PETR.ZS_DS2_en_csv_v2_4356435.json'
production_by_renewable =  path+'electricity_production_renewable_sources/API_EG.ELC.RNWX.KH_DS2_en_csv_v2_4356007.json'
```

## Reading the JSON files

```python
access_to_rural_1= pd.read_json(access_to_rural)
access_to_urban_1 = pd.read_json(access_to_urban)
access_to_world_1 = pd.read_json(access_to_world)
transmission_distribution_losses_1 =pd.read_json(transmission_distribution_losses)
production_by_nuclear_1 = pd.read_json(production_by_nuclear)
production_by_oil_1 = pd.read_json(production_by_oil)
production_by_renewable_1 = pd.read_json(production_by_renewable)
```

## All Data from JSON file

```python
df_data = [access_to_rural_1,access_to_urban_1,access_to_world_1,transmission_distribution_losses_1,
           production_by_nuclear_1,production_by_oil_1,production_by_renewable_1]
```

```python
num_col=[]
for i in range(1960,2022):
    num_col.append(i)
cat_col=['Country_Name','Country_Code','Indicator_Name','Indicator_Code']
col = cat_col + num_col
```

## Dropping Unnesscary Columns and Values

```python
for df in df_data:
    df.drop(0,axis  = 0 , inplace = True)
    df.drop(66, axis = 1 , inplace = True)
    df.columns = col
    df.drop(1,axis = 0 , inplace = True)
    df.reset_index(inplace = True)
    df.drop('index',axis = 1 ,inplace = True)
```

```python
df_data[1]
```

## Changing the relevant datatypes

```python
for i in num_col:
    df[i] = pd.to_numeric(df[i],errors='coerce')
```

## Null Values To Zero  

```python
for df in df_data:
    df.fillna(0,inplace=True)
```

## Stacking Up Year Column into single Column

```python
for i in range(0,7):
    df_data[i] = df_data[i][cat_col].merge(df_data[i].set_index('Country_Code')[num_col].stack().reset_index(),on='Country_Code')
    df_data[i] = df_data[i].rename(columns = {'level_1':'Year',0:'d_point'})
```
